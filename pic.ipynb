```python
import numpy as np
import pandas as pd
from data_pre import data_normalize
import matplotlib.pyplot as plt
```

```python
wd = pd.DataFrame(columns=['WD'], index=df['WD'].unique().tolist())
for i in df['WD'].unique():
    wd.loc[i, 'WD'] = df['WD'].tolist().count(i)
wd.plot.bar()
```

```python
na = 'WLA'
wla = []
k = 0
wla_ = []
while k < 27:
    k += 3
    wla.append(len(df[(df[na] < k) & (df[na] >= k - 3)][na].tolist()))
    wla_.append('{}-{}'.format(k-3, k))
df_wla = pd.DataFrame(columns=['WLA'], index=wla_)
df_wla['WLA'] = wla
df_wla.plot.bar()
```

```python
na = 'WIA'
wia = []
k = 0
wia_ = []
while k < 13:
    k += 1
    wia.append(len(df[(df[na] < k) & (df[na] >= k - 1)][na].tolist()))
    wia_.append('{}-{}'.format(k-1, k))
df_wia = pd.DataFrame(columns=['WIA'], index=wia_)
df_wia['WIA'] = wia
df_wia.plot.bar()
```

```python
na = 'TD'
td = []
k = 0
td_ = []
while k < 110:
    k += 10
    td.append(len(df[(df[na] < k) & (df[na] >= k - 10)][na].tolist()))
    td_.append('{}-{}'.format(k-10, k))
df_td = pd.DataFrame(columns=['TD'], index=td_)
df_td['TD'] = td
df_td
```

```python
na = 'TI'
ti = []
ti_ = []
k = 0
while k < 30:
    k += 3
    ti.append(len(df[(df[na] < k) & (df[na] >= k - 3)][na].tolist()))
    ti_.append('{}-{}'.format(k-3, k))
df_ti = pd.DataFrame(index=ti_, columns=['TI'])
df_ti['TI'] = ti
df_ti
```

```python
na = 'UCS'
ucs = []
k = 0
ucs_ = []
while k < 70:
    k += 7
    ucs.append(len(df[(df[na] < k) & (df[na] >= k - 7)][na].tolist()))
    ucs_.append('{}-{}'.format(k-7, k))
df_ucs = pd.DataFrame(columns=['UCS'], index=ucs_)
df_ucs['UCS']=ucs
df_ucs
```

```python
na = 'MTL'
mtl = []
k = 0
mtl_ = []
while k < 2:
    k += 0.2
    mtl.append(len(df[(df[na] < k) & (df[na] >= k - 0.2)][na].tolist()))
    mtl_.append('{}-{}'.format(round(k-0.2, 1), round(k, 1)))
df_mtl = pd.DataFrame(columns=['MTL'], index=mtl_)
df_mtl['MTL']=mtl
df_mtl
```

```python
tn = pd.DataFrame(columns=['TN'], index=df['TN'].unique().tolist())
for i in df['TN'].unique():
    tn.loc[i, 'TN'] = df['TN'].tolist().count(i)
tn
```

```python
na = 'ATI'
ati = []
k = 0
ati_ = []
while k < 0.75:
    k += 0.15
    ati.append(len(df[(df[na] < k) & (df[na] >= k - 0.15)][na].tolist()))
    ati_.append('{}-{}'.format(round(k-0.15, 2), round(k, 2)))
df_ati = pd.DataFrame(columns=['ATI'], index=ati_)
df_ati['ATI']=ati
df_ati
```

```python
na = 'TBD'
tbd = []
k = 0
tbd_ = []
while k < 400:
    k += 50
    tbd.append(len(df[(df[na] < k) & (df[na] >= k - 50)][na].tolist()))
    tbd_.append('{}-{}'.format(k-50, k))
df_tbd = pd.DataFrame(columns=['TBD'], index=tbd_)
df_tbd['TBD']=tbd
df_tbd
```

```python
from matplotlib.ticker import FormatStrFormatter
import seaborn as sns
import warnings
# sns.set_theme(style='white')
import math
warnings.filterwarnings('ignore')
fig, axs = plt.subplots(3, 5, figsize=(30, 15))

n = df.columns[1:]
fig.subplots_adjust(wspace=0.25, hspace=0.2)
plt.rcParams['font.sans-serif'] = ['Times New Roman']
plt.rcParams['axes.unicode_minus'] =False
plt.rcParams['font.size'] = 18
plt.rc('xtick', labelsize=18)  # 刻度标签大小
plt.rc('ytick', labelsize=18)

for ax in [axs[0, i] for i in range(5) if i != 2]:
    ax.set_visible(False)

sns.histplot(df.iloc[:, 0], ax=axs[0, 2], palette=["steelblue"],
             edgecolor='w')
axs[0, 2].grid()
ax = axs[0, 2].twinx()
axs[0, 2].set_ylabel('Count', weight='bold', fontsize=18)
ax.set_ylabel('Density', weight='bold', fontsize=18)
u = np.average(df.iloc[:, 0])
sig = np.sqrt(np.var(df.iloc[:, 0], ddof=1))
x = np.linspace(0, u + 3*sig, 100)   # 定义域
y = np.exp(-(x - u) ** 2 / (2 * sig ** 2)) / (np.sqrt(2*np.pi)*sig) # 定义曲线函数
ax.plot(x, y, c='r', label='Normal')
plt.legend(fontsize=18, loc='lower left')
axis = axs[0, 2].axis()
axs[0, 2].set_xlim(0, 5.5)
axs[0, 2].set_ylim(axis[2], axis[3])
message = df.describe().iloc[:, 0]
axs[0, 2].text(axis[1] * 11 / 20, axis[3] * 2 / 3, 'n = {}\nMin = {}\nMax = {}\nMean = {}\nStd.Dev = {}'.format(message['count'],
                       message['min'], 
                       message['max'], 
                       round(message['mean'], 2), 
                       round(message['std'], 2)),
                       weight='bold',
                       fontsize=16)

for i in range(2):
    for j in range(5):
        sns.histplot(df[n[5 * i + j]], ax=axs[i + 1, j], 
                     palette=["steelblue"],
                     edgecolor='w')
        axs[i + 1, j].grid()
        # axs[i, j].set_xticks(ticks=axs[i, j].ticks, fontsize=16)
        ax = axs[i + 1, j].twinx()
        """ sns.kdeplot(df[n[5 * i + j]], ax=ax, color='r', linewidth=2.4, legend=False) """
        axs[i + 1, j].set_xlabel(n[5 * i + j], weight='bold', fontsize=18)
        if j == 0:
            axs[i + 1, j].set_ylabel('Count', weight='bold', fontsize=18)
        else:
            axs[i + 1, j].set_ylabel('', weight='bold', fontsize=18)
        if j == 4:
            ax.set_ylabel('Density', weight='bold', fontsize=18)
        else:
            ax.set_ylabel('', weight='bold', fontsize=18)
        y_formatter = FormatStrFormatter('%1.2f')
        ax.yaxis.set_major_formatter(y_formatter)
        axis = axs[i + 1, j].axis()
        message = df.describe()[n[5 * i + j]]
        # axs[i, j].plot(x, y)
        if ((j == 1) | (j == 2)) and (i == 1):
            axs[i + 1, j].text(axis[1] * 1 / 20, axis[3] * 2 / 3, 'n = {}\nMin = {}\nMax = {}\nMean = {}\nStd.Dev = {}'.format(message['count'],
                       message['min'], 
                       message['max'], 
                       round(message['mean'], 2), 
                       round(message['std'], 2)), 
                       fontsize=16,
                       weight='bold')
        else:
            axs[i + 1, j].text(axis[1] * 3 / 5, axis[3] * 2 / 3, 'n = {}\nMin = {}\nMax = {}\nMean = {}\nStd.Dev = {}'.format(message['count'],
                       message['min'], 
                       message['max'], 
                       round(message['mean'], 2), 
                       round(message['std'], 2)),
                       weight='bold',
                       fontsize=16)
        
        """正态分布曲线"""
        u = np.average(df[n[5 * i + j]])
        sig = np.sqrt(np.var(df[n[5 * i + j]], ddof=1))
        x = np.linspace(0, u + 3*sig, 100)   # 定义域
        y = np.exp(-(x - u) ** 2 / (2 * sig ** 2)) / (np.sqrt(2*np.pi)*sig) # 定义曲线函数
        ax.plot(x, y, c='r', label='Normal')
        plt.legend(fontsize=18, loc='lower left')
        axs[i + 1, j].set_xlim(0, axis[1])
        axs[i+1, j].set_ylim(axis[2], axis[3])
```
